# Problem 2: German Credit Dataset

This notebook performs preprocessing, analysis, and visualization on the German Credit dataset as required in the homework assignment.


## Task A1 (8 pts): Drop the 3 columns that contain the highest number of 'none' values.

We count occurrences of the string `none` in each column, break ties left-to-right, and drop the top 3.


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("GermanCredit.csv")

# Count 'none' occurrences
none_counts = {col: (df[col] == 'none').sum() for col in df.columns}

# Sort by count desc, tie-break by column index
sorted_cols = sorted(
    none_counts.items(),
    key=lambda x: (-x[1], list(df.columns).index(x[0]))
)

cols_to_drop = [col for col, count in sorted_cols[:3]]

df.drop(columns=cols_to_drop, inplace=True)

df.head()
```

## Task A2 (4 pts): Remove all apostrophes from string columns.


```python
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].str.replace("'", "", regex=False)
```

## Task A3 (5 pts): Remap checking_status categories.


```python
checking_map = {
  "no checking": "No Checking",
  "<0": "Low",
  "0<=X<200": "Medium",
  ">=200": "High",
};
df["checking_status"] = df["checking_status"].map(checking_map);
```

## Task A4 (5 pts): Remap savings_status categories.


```python
savings_map = {
    'no known savings': 'No Savings',
    '<100': 'Low',
    '100<=X<500': 'Medium',
    '500<=X<1000': 'High',
    '>=1000': 'High'
}
df['savings_status'] = df['savings_status'].map(savings_map)

```

## Task A5 (3 pts): Convert class column ('good' → 1, 'bad' → 0).


```python
df['class'] = df['class'].map({'good': 1, 'bad': 0})
```

## Task A6 (4 pts): Remap employment categories.


```python
def map_employment(x):
    x = x.strip()
    if x == 'unemployed':
        return 'Unemployed'
    if x == '<1':
        return 'Amateur'
    if x == '1<=X<4':
        return 'Professional'
    if x == '4<=X<7':
        return 'Experienced'
    if x == '>=7':
        return 'Expert'

df['employment'] = df['employment'].apply(map_employment)
```

## Task B1 (5 pts): Crosstab — foreign_worker vs class


```python
pd.crosstab(df['foreign_worker'], df['class'])
```

## Task B2 (2 pts): Crosstab — employment vs savings_status


```python
pd.crosstab(df['employment'], df['savings_status'])
```

## Task B3 (4 pts): Average credit_amount of single males with 4–7 years employment.

This corresponds to:

- personal_status = 'male single'
- employment = 'Experienced'


```python
mask = (df['personal_status'] == 'male single') & (df['employment'] == 'Experienced')
df.loc[mask, 'credit_amount'].mean()
```

## Task B4 (4 pts): Average credit duration for each job type.


```python
df.groupby('job')['duration'].mean()
```

## Task B5 (4 pts): Most common checking_status and savings_status for purpose='education'


```python
edu = df[df['purpose'] == 'education']

print("Most common checking status:", edu['checking_status'].mode()[0])
print("Most common savings status:", edu['savings_status'].mode()[0])
```

## Task C1 (9 pts): Subplots of two bar charts:

1. savings_status → personal_status
2. checking_status → personal_status


```python
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

df.groupby(['savings_status', 'personal_status']).size().unstack().plot(
    kind='bar', ax=axes[0], title="Savings Status → Personal Status"
)

df.groupby(['checking_status', 'personal_status']).size().unstack().plot(
    kind='bar', ax=axes[1], title="Checking Status → Personal Status"
)

plt.tight_layout()
plt.show()
```

## Task C2 (9 pts): For credit_amount > 4000, bar chart of property_magnitude vs average age.


```python
filtered = df[df['credit_amount'] > 4000]
avg_age = filtered.groupby('property_magnitude')['age'].mean()

avg_age.plot(kind='bar', figsize=(8,5), title="Avg Age vs Property Magnitude (credit > 4000)")
plt.ylabel("Average Age")
plt.show()
```

## Task C3 (6 pts): Pie charts for High savings_status and age > 40:

- Personal status
- Credit history
- Job


```python
subset = df[(df['savings_status'] == 'High') & (df['age'] > 40)]

fig, axes = plt.subplots(1, 3, figsize=(18, 6))

subset['personal_status'].value_counts().plot.pie(
    ax=axes[0], autopct='%1.1f%%', title="Personal Status")

subset['credit_history'].value_counts().plot.pie(
    ax=axes[1], autopct='%1.1f%%', title="Credit History")

subset['job'].value_counts().plot.pie(
    ax=axes[2], autopct='%1.1f%%', title="Job")

plt.show()
```
